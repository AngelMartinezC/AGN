## Program to read fits spectrum file of an AGN

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

### Leer espectros como datos hdu

```python
hdudata = fits.getdata("spec-2127-53859-0085.fits")
hduheader = fits.getheader("spec-2127-53859-0085.fits")
```

```python
# Convertir los datos a lista
hdudata2 = [i for i in hdudata]
```

```python
# Convertir la lista a array de numpy y transponerla para obtener
# la información de flujo [0], log10 de la longitud de onda [1]
# y demás datos

hdudata3 = np.transpose(np.array(hdudata2))
```

```python
x_axis = 10**hdudata3[1]    # Longitud de onda, segunda columna
y_axis = hdudata3[0]        # Flujo, primera columna
```

```python
# Plot figure
%matplotlib inline
plt.figure(figsize=(10,6))
plt.plot(x_axis,y_axis)
plt.xlabel(r"Wavelength [$\AA$]")
plt.ylabel(r"Flux [$10^{-7}$erg s$^{-1}$ cm$^{-2}$ $\AA^{-1}$]")
plt.grid()
```
